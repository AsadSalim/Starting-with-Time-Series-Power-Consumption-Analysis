### Feature Creation

```python
def create_features(df):
    """ 
    Create time series features based on time series index.
    """
    df = df.copy()
    df['hour'] = df.index.hour
    df['dayofweek'] = df.index.dayofweek
    df['quarter'] = df.index.quarter
    df['month'] = df.index.month
    df['year'] = df.index.year
    df['dayofyear'] = df.index.dayofyear
    return df
df = create_features(df)
```

```python
df
```

### Visualize Feature / Target Relationship

```python
fig, ax = plt.subplots(figsize = (10, 8))
sns.boxplot(data = df, x = 'hour', y = 'PJME_MW')
ax.set_title('MW per hour')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (10, 8))
sns.boxplot(data = df, x = 'month', y = 'PJME_MW', palette = 'Greens')
ax.set_title('MW per month')
plt.show()
```

### Create our Model

```python
train = create_features(train)
test = create_features(test)
```

```python
train
```

```python
# Let's define our features 
df.columns
```

```python
FEATURES = ['hour', 'dayofweek', 'quarter', 'month', 'year',
       'dayofyear']
TARGET = 'PJME_MW'
```

```python
X_train = train[FEATURES]
y_train = train[TARGET]

X_test = test[FEATURES]
y_test = test[TARGET]
```

```python
reg = xgb.XGBRegressor(n_estimators = 1000, early_stopping_rounds = 50, learning_rate = 0.01)
reg.fit(X_train, y_train, eval_set = [(X_train, y_train), (X_test, y_test)], verbose = 100)
```

### Feature Importance

```python
fi = pd.DataFrame(data = reg.feature_importances_*100, 
                  index = reg.feature_names_in_,
                 columns = ['Importance'])
```

```python
fi.sort_values('Importance').plot(kind = 'barh', title = 'Feature_Importance')
plt.show()
```

### Forecast on test set

```python
test['prediction'] = reg.predict(X_test)
df = df.merge(test[['prediction']], how = 'left', left_index = True, right_index = True)
ax = df[['PJME_MW']].plot(figsize = (15, 5))
df['prediction'].plot(ax = ax, style = '.')
plt.legend(['Truth Data', 'Predictions'])
ax.set_title('Raw Data and Predictions')
plt.show()
```

```python
# Let's look at a week prediction
ax = df.loc[(df.index > '04-01-2018') & (df.index < '04-08-2018')]['PJME_MW'].plot(figsize = (15, 5), title = 'Week of Data')
df.loc[(df.index > '04-01-2018') & (df.index < '04-08-2018')]['prediction'].plot(ax = ax, style = '.')
plt.legend(['Truth Data', 'Predictions'])
ax.set_title('Raw Data and Predictions')
plt.show()
```

```python
# The model did not make a perfect job and some improvement can be made including
# 1- Parameter tuning
# 2- Adding more relevent features
# 3- Robust cross validation
```

### Running our evaluation metric on this

```python
# Let's evaluate this model
score = np.sqrt(mean_squared_error(test['PJME_MW'], test['prediction']))
print(f'RMSE score on test set: {score:0.2f}')
```

### Calculate error

```python
# Looking at the best and worst predicted days
test['error'] = np.abs(test[TARGET] - test['prediction'])
```

```python
test['date'] = test.index.date
```

```python
# Worst predicted values
test.groupby('date')['error'].mean().sort_values(ascending = False).head(5)
```
